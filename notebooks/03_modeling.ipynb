```python
DATA_PATH = "../data/Processed/V3Processed.csv"
```

```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from category_encoders import BinaryEncoder
```

```python
df = pd.read_pickle(DATA_PATH)
```

```python
df.info()
```

```python
# Save the state of NaN values before encoding
nan_before_encoding = df.isna().sum()
print("NaN values before encoding:\n", nan_before_encoding[nan_before_encoding > 0])
```

```python
df['property'].unique()
```

```python
dct1={
    'Single Family' : 1 ,
    'Two Family' : 2,
    'Three Family' : 3,
    'Four Family' : 4 }
```

```python
df['property_encoded_map'] = df['property'].map(dct1)
df.drop(columns=['property'], inplace=True)
```

```python
df['residential'].unique()
```

```python
dct2={
    'Detached House' : 1 ,
    'Duplex' : 2,
    'Triplex' : 3,
    'Fourplex' : 4 }
```

```python
df['residential_encoded_map'] = df['residential'].map(dct2)
df.drop(columns=['residential'], inplace=True)
```

```python
df['month_name'].unique()
```

```python
dct3={
    'January' : 1 ,
    'February' : 2,
    'March' : 3,
    'April' : 4 ,
    'May' : 5,
    'June' :6 ,
    'July' : 7,
    'August' : 8,
    'September' : 9,
    'October' : 10 ,
    'November' : 11 ,
    'December' : 12 ,    
    }
```

```python
df['month_name_encoded_map'] = df['month_name'].map(dct3)
df.drop(columns=['month_name'], inplace=True)
```

```python
df['locality'].unique()
```

```python
be=BinaryEncoder()
be_df=be.fit_transform(df['locality'])
df=pd.concat([df,be_df] , axis=1)
df.drop(['locality'] , axis=1, inplace=True)
```

```python
df['face'].unique()
```

```python
be=BinaryEncoder()
be_df=be.fit_transform(df['face'])
df=pd.concat([df,be_df] , axis=1)
df.drop(['face'] , axis=1, inplace=True)
```
